```python
import pandas as pd
from scipy import stats
import numpy as np

#upload csv data
from google.colab import files
uploaded = files.upload()
```

```python
#load data and display
df= pd.read_csv('customer_demographics.csv')
df.head(10)
```

```python
# Convert 'family_size' to numeric, coercing errors to NaN (non-numeric values become NaN)
df['family_size'] = pd.to_numeric(df['family_size'], errors='coerce')

# Convert 'age_range' to string, then extract numeric part
df['age_range'] = df['age_range'].astype(str).str.extract('(\d+)').astype(float)

# Drop rows with NaN values in the columns of interest
df = df.dropna(subset=['family_size', 'age_range'])

# Columns to analyze
columns_to_check = ['family_size', 'age_range']

# Function to analyze data
def analyze_column(column):
    data = df[column].dropna()

    # Shapiro-Wilk test
    shapiro_test = stats.shapiro(data)

    # Chi-square test of goodness-of-fit
    observed_freq, bin_edges = np.histogram(data, bins='auto')
    # Ensure that the expected frequencies have the same number of bins
    expected_freq = np.ones_like(observed_freq) * len(data) / len(observed_freq)
    chi2_stat, p_value = stats.chisquare(observed_freq, expected_freq)

    # Skewness
    skewness = stats.skew(data)

    # Kurtosis
    kurtosis = stats.kurtosis(data)

    # Standard deviation
    sigma = np.std(data, ddof=1)

    return {
        'Shapiro-Wilk p-value': shapiro_test.pvalue,
        'Chi-square p-value': p_value,
        'Skewness': skewness,
        'Kurtosis': kurtosis,
        'Standard Deviation': sigma
    }

# Analyze each column
results = {}
for column in columns_to_check:
    results[column] = analyze_column(column)

# Print results
for column, metrics in results.items():
    print(f'\nAnalysis for column: {column}')
    for metric, value in metrics.items():
        print(f'{metric}: {value}')
```
