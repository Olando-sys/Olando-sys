```python
# Install required libraries
!pip install requests beautifulsoup4 pandas

import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the Wikipedia page
url = 'https://en.wikipedia.org/wiki/List_of_cities_in_Kenya'

# Send a request to fetch the content of the page
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Find all tables
tables = soup.find_all('table')

# Check the number of tables found
print(f"Number of tables found: {len(tables)}")

# Preview each table to identify the correct one
for i, table in enumerate(tables):
    print(f"\nPreview of table {i}:")
    rows = table.find_all('tr')
    for row in rows[:5]:
        print([cell.get_text(strip=True) for cell in row.find_all('td')])


table_index = 0
table = tables[table_index]

# Initialize lists to store data
cities = []

# Iterate over table rows to extract city names
for row in table.find_all('tr')[1:]:
    columns = row.find_all('td')
    if columns:
        city = columns[0].get_text(strip=True)
        cities.append(city)

# Create a DataFrame from the list of cities
df = pd.DataFrame(cities, columns=['City'])

# Display the DataFrame
df.head()
```
