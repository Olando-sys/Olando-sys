```python
#Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Load the Iris dataset from sklearn
iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# Display the first few rows of the dataset
iris_df.head()
```

```python
# Load the Iris dataset from sklearn
iris_sklearn = load_iris()

# Convert to DataFrame
iris_df = pd.DataFrame(data=iris_sklearn.data, columns=iris_sklearn.feature_names)
iris_df['species'] = iris_sklearn.target

# Map species names
species_mapping = {0: 'setosa', 1: 'versicolor', 2: 'virginica'}
iris_df['species'] = iris_df['species'].map(species_mapping)

# Display the first few rows of the dataset
iris_df.head()
```

```python
# Basic statistical characteristics
basic_stats = iris_df.describe()

# Checking for missing values
missing_values = iris_df.isnull().sum()

basic_stats, missing_values
```

```python
#Set the aesthetic style of the plots
sns.set(style="whitegrid")

# 1. Pair Plot of Iris Species
pair_plot = sns.pairplot(iris_df, hue='species', markers=["o", "s", "D"], palette="Set2")
pair_plot.fig.suptitle("Pair Plot of Iris Species", y=1.02)  # Adjusting the title position

# 2. Sepal Length Distribution by Species using boxplot
plt.figure(figsize=(10, 6))
box_plot = sns.boxplot(x='species', y='sepal length (cm)', data=iris_df, palette="Set2")
box_plot.set_title("Sepal Length Distribution by Species")
box_plot.set_xlabel("Species")
box_plot.set_ylabel("Sepal Length (cm)")

# Show plots
plt.show()
```

```python
#Calculate the correlation matrix for numerical columns only
corr_matrix = iris_df.drop('species', axis=1).corr()

# Display the correlation matrix
corr_matrix

# Create a heatmap
plt.figure(figsize=(8, 6))
heatmap = sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
heatmap.set_title("Correlation Heatmap of Iris Features")
plt.show()
```
